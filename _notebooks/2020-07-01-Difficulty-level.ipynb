```
#hide

aula1 = pd.read_csv('Feedback - Class 01-Grid view.csv')
aula1['Aula'] = 1
aula2 = pd.read_csv('Feedback - Class 02-Grid view.csv')
aula2['Aula'] = 2
aula3 = pd.read_csv('Feedback - Class 03-Grid view.csv')
aula3['Aula'] = 3
aula4 = pd.read_csv('Feedback - Class 04-Grid view.csv')
aula4['Aula'] = 4
aula5 = pd.read_csv('Feedback - Class 05-Grid view.csv')
aula5['Aula'] = 5
aula6 = pd.read_csv('Feedback - Class 06-Grid view.csv')
aula6['Aula'] = 6
```

```

#hide
aulas = pd.concat([aula1,aula2,aula3,aula4,aula5,aula6])
```

```
#hide

sem_1_e_2 = aulas[aulas['Aula']>2]
sem_1_e_2=sem_1_e_2.dropna(subset=['Avalie a dificuldade da aula'])
sem_1_e_2['Avalie a dificuldade da aula'] =sem_1_e_2['Avalie a dificuldade da aula'].apply(lambda x: str(x))
sem_1_e_2['Avalie a dificuldade da aula'] =sem_1_e_2['Avalie a dificuldade da aula'].apply(lambda x: x.replace(',','.'))
sem_1_e_2['Avalie a dificuldade da aula'] =pd.to_numeric(sem_1_e_2['Avalie a dificuldade da aula'])
```

```
#hide

grouped = sem_1_e_2.groupby('Aula').agg('mean')
grouped
```

```
#hide

grouped.plot(kind='bar')
```
