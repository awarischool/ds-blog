```
#hide
import pandas as pd
import numpy as np
import os
import plotly.express as px
```

```
#hide

os.chdir('drive/My Drive/Feedbacks/')
```

```
#hide

aula1 = pd.read_csv('Feedback - Class 01-Grid view.csv')
aula1['Aula'] = 1
aula2 = pd.read_csv('Feedback - Class 02-Grid view.csv')
aula2['Aula'] = 2
aula3 = pd.read_csv('Feedback - Class 03-Grid view.csv')
aula3['Aula'] = 3
aula4 = pd.read_csv('Feedback - Class 04-Grid view.csv')
aula4['Aula'] = 4
aula5 = pd.read_csv('Feedback - Class 05-Grid view.csv')
aula5['Aula'] = 5
aula6 = pd.read_csv('Feedback - Class 06-Grid view.csv')
aula6['Aula'] = 6
aula7 = pd.read_csv('Feedback - Class 07-Grid view.csv')
aula7['Aula'] = 7
```

```

#hide
aulas = pd.concat([aula1,aula2,aula3,aula4,aula5,aula6,aula7])
```

```
#hide

sem_1_e_2 = aulas[aulas['Aula']>2]
sem_1_e_2=sem_1_e_2.dropna(subset=['Avalie a dificuldade da aula'])
sem_1_e_2['Avalie a dificuldade da aula'] =sem_1_e_2['Avalie a dificuldade da aula'].apply(lambda x: str(x))
sem_1_e_2['Avalie a dificuldade da aula'] =sem_1_e_2['Avalie a dificuldade da aula'].apply(lambda x: x.replace(',','.'))
sem_1_e_2['Avalie a dificuldade da aula'] =pd.to_numeric(sem_1_e_2['Avalie a dificuldade da aula'])
```

```
#hide

grouped = sem_1_e_2.groupby('Aula').agg('mean')
grouped
```

```
#hide_input
fig = px.bar(data_frame=grouped,x=grouped.index,y='Avalie a dificuldade da aula',
             title="Nível de dificuldade das aulas de acordo com os alunos",labels = {'x':'Nº da Aula','Avalie a dificuldade da aula':'Dificuldade'})             
fig.update_layout(yaxis=dict(range=[3,10],dtick=1))
fig.update_layout(xaxis=dict(range=[2,8],dtick=1))

fig.show()
```

```
#hide_input
fig = px.bar(data_frame=grouped,x=grouped.index,y='Avalie o Material Disponibilizado',
             title="Qualidade do material de acordo com os alunos",labels = {'x':'Nº da Aula','Avalie o Material Disponibilizado':'Material'})
fig.update_layout(yaxis=dict(range=[0,5],dtick=1))
fig.update_layout(xaxis=dict(range=[2,8],dtick=1))

fig.show()
```

```
#hide_input
fig = px.bar(data_frame=grouped,x=grouped.index,y='Avalie a Didática e Clareza da Aula',
             title="Qualidade da didatática das aulas de acordo com os alunos",labels = {'x':'Nº da Aula','Avalie a Didática e Clareza da Aula':'Didática'})
fig.update_layout(yaxis=dict(range=[0,5],dtick=1))
fig.update_layout(xaxis=dict(range=[2,8],dtick=1))

fig.show()
```
